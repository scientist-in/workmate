```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =[i[0] for i in acc_int]
y =[i[1] for i in acc_int]
z =[i[2] for i in acc_int]



ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
#clustering

from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array(acc_int)
```

```python
X
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(X)
```

```python
kmeans.labels_
#array([0, 0, 0, 1, 1, 1], dtype=int32)
```

```python
#kmeans.predict([[0, 0], [4, 4]])
#array([0, 1], dtype=int32)
```

```python
clusters = kmeans.cluster_centers_
#array([[ 1.,  2.],
#       [ 4.,  2.]])
```

```python
clusters
```

```python
#find face for all entries
face_recording = [np.argmin([np.linalg.norm(face_entry- i )for i in clusters]) for face_entry in acc_int]
```

```python
time_face = zip(time_fmt,face_recording)
with open('time_face2.csv','wb') as thefile:
    for item in time_face:
        thefile.write("{0}\n".format(item))
```
